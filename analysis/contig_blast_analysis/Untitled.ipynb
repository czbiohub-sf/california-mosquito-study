```python
import pandas as pd
```

```python
bony_fish_df = pd.read_csv("untitled.txt", delim_whitespace=True, header=None, names=["qseqid", "sseqid", "pident", "length", "mismatch", "gapopen", 
                   "qstart", "qend", "sstart", "send", "evalue", "bitscore"])

```

```python
bony_fish_df
```

```python
bony_fish_df = bony_fish_df.assign(blast_type="rapsearch")
bony_fish_df = bony_fish_df.assign(qlength=668)
```

```python
bony_fish_df = bony_fish_df.assign(qcov=abs(bony_fish_df.qstart-bony_fish_df.qend)+(1-bony_fish_df.mismatch))#*[1 if x=="gsnap" else 3 for x in bony_fish_df.blast_type])
```

```python
bony_fish_df = bony_fish_df.assign(qcov_prop=bony_fish_df.qcov/bony_fish_df.qlength)
```

```python
bony_fish_df = bony_fish_df.assign(pmatch=bony_fish_df.qcov_prop * bony_fish_df.pident / 100)
```

```python
def get_pmatch_per_contig (df):
    if (len(df.index)==1):
        return (df["pmatch"].iloc[0])
    covered_bases = []
    for row_i in range(len(df.index)):
        start_end = df.iloc[row_i, :][["qend", "qstart"]].sort_values()
        covered_bases += list(range(start_end[0], start_end[1]))
    total_mismatch = sum(df["mismatch"]) * [1 if x=="gsnap" else 3 for x in df.blast_type][0]
    print(total_mismatch)
    print (total_mismatch/3)
    return(float((len(set(covered_bases))-total_mismatch)/df.head(1)["qlength"]))
```
