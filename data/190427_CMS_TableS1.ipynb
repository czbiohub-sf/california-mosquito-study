#Generating new dataframe "TableS1data" with subset of relevant data for TableS1.

```python
TableS1data = AllMosData.loc[:,('NewIDseqName','compute_genus','compute_species','collected_by','collection_date','collection_lat','collection_long','Habitat')]
```

```python
TableS1data
```

```python
TableS1data.sort_values(by = ['collected_by','compute_genus','compute_species'])
```

#Strip unnecessary suffix string from 'NewIDseqName' values

```python
TableS1data.replace(r'\_Ra.*', '', regex=True)
```

#Strip unnecessary time suffix string from 'collection_date' values

```python
TableS1data.replace(r'T00.*', '', regex=True)
```

#Tidying up dataframe: renaming columns

```python
TableS1data.rename(columns={'NewIDseqName':'Specimen ID','compute_genus':'Genus','compute_species':'Species','collected_by':'Collection Site','collection_date':'Collection Date','collection_lat':'Latitude','collection_long':'Longitude'}, inplace=True)
```

```python
TableS1data.head()
```

```python
TableS1data[['Specimen ID','Genus','Species','Collection Site','Collection Date','Latitude','Longitude','Habitat']].head()
```

```python
with open('TableS1data.tex','w') as tf:
    tf.write(TableS1data.to_latex())
```

#Playing with table formatting - specifying column alignment, etc.

#Using style to eliminate index, align text.

```python
TableS1data.style.hide_index().set_properties(**{'text-align': 'left'})
```

#More sophisticated (props to StackOverflow): specifying multiple items - hide index, specify header and cell alignments.

```python
styler = TableS1data.style
#hide the index values
styler = styler.hide_index()
#left align cell values
styler = styler.set_properties(**{'text-align': 'left'})
# select the table header with th and set it left align
styler.set_table_styles([dict(selector="th", props=[("text-align", "left")])])
```
