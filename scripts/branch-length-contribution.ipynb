```python
import importlib.machinery
import importlib.util
loader = importlib.machinery.SourceFileLoader('baltic','/Users/evogytis/Documents/baltic/baltic.py')
spec = importlib.util.spec_from_loader(loader.name, loader)
bt = importlib.util.module_from_spec(spec)
loader.exec_module(bt)

import re,glob,os,json
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import matplotlib.patheffects as path_effects
from matplotlib.gridspec import GridSpec

typeface='Helvetica Neue'
# mpl.rcParams['font.weight']=300
# mpl.rcParams['axes.labelweight']=300
mpl.rcParams['font.family']=typeface
mpl.rcParams['font.size']=22

def dist2root(node):
    cur_node=node
    lvl=0
    while cur_node:
        cur_node=cur_node.parent
        lvl+=1
        
    return lvl

new_virus={}
pol2contigs={}
virus=json.load(open('/Users/evogytis/Documents/manuscripts/skeeters/data/darkmatter/virus.json','r'))
for pol in virus:
    pol2contigs[pol]=[]
    if 'provisional_name' in virus[pol]:
        new_virus[pol]=True
    else:
        new_virus[pol]=False
    for seg in virus[pol]['segments']:
        for contig in virus[pol]['segments'][seg]['contigs']:
            pol2contigs[pol].append(contig)

pol2colour={}
for line in open('/Users/evogytis/Documents/manuscripts/skeeters/figures/fig3/virus_color_scheme.tsv','r'):
    l=line.strip('\n').split('\t')
    pol2colour[l[0]]=l[-1]
            
base_path='/Users/evogytis/Documents/manuscripts/skeeters/data/trees/fig7_branchLength/'

Xs={}
Ys={}
Cs={}
```

```python
taxonomy_trees={'Mononega-Chu': '((("Xinmoviridae"[&subtree="Xinmo"],("Nyamiviridae","Bornaviridae")),(((("Paramyxoviridae","Filoviridae"),"Novirhabdovirus"),"unnamed group 1"[&subtree="Rhabdo-like"]),(("unnamed group 2"[&subtree="CytoNucleorhabdo-like"],("Cytorhabdovirus","Nucleorhabdovirus")),"Dimarhabdovirus"[&subtree="Dimarhabdo"]))),"Chuviridae"[&subtree="Chu"]);', 
                'Bunya': '(("Phasmavirus"[&subtree="PhasmaFeraJon"],("Hantavirus",("Orthobunyavirus"[&subtree="Orthobunya"],("Emaravirus","Tospovirus")))),("Phlebovirus"[&subtree="Phlebo-like"],("Nairovirus","Arenavirus")));', 
                'NarnaLevi': '("Leviviridae"[&subtree="Levi"],("Mitovirus",("Botourmiaviridae"[&subtree="Botourmia"],("Narnavirus"[&subtree="Narna"],"Narna-like group"[&subtree="Narna-like"]))));', 
                'Partiti': '(("Picobirna","Picobirna-like"),(("Gammapartitivirus","Deltapartitivirus"),("Amalgaviridae",(("Alphapartitivirus","Betapartitivirus"),"Epsilonpartitivirus"[&subtree="Partiti-like"]))));', 
                'Flavi': '(("Pestivirus"[&subtree="Pesti-like"],("Pegivirus","Hepacivirus")),("Flavivirus"[&subtree="Flavi-like"],"Jingmenvirus"));', 
                'Solemo': '(("La Tardoire-like group"[&subtree="LaTardoire-like"],"Motts Mills-like group"[&subtree="MottsMills-like"]),(("Luteoviridae"[&subtree="Polero-like"],"Sobemovirus"),"Sobemo-like groups"));', 
                'Reo': '(((("Orbivirus","Orbivirus-like"[&subtree="Orbi-like"]),"Mimoreovirus"),("Rotavirus",("Seadornavirus","Phytoreovirus"))),("Orthoreovirus",("unnamed group 1",("Colti-Mycovirus",("Fijivirus",("Oryzavirus","Cypovirus"[&subtree="Cypo"]))))));', 
                'Tombus': '("Noda-like",("Lake Sinai-like group"[&subtree="LakeSinai-like"],("Jingmen tombus 2-like group"[&subtree="JingmenTombus2-like"],("Beihai tombus 15-like groups",("Diaphorina citri associated C-like group",("Corseley-like group"[&subtree="Corseley-like"],"Tombusviridae")))));', 
                'Toti': '((("Piscine myocarditis-like group","Camponotus nipponicus-like group"[&subtree="Camponotus-nipponicus-like"]),"Omono river-like group"[&subtree="Omono-river-like"]),("Totivirus"[&subtree="Totivirus"],("Chrysoviridae"[&subtree="Chryso"],((("Victorivirus","Leishamniavirus"),"Trichomonasvirus")))));', 
                'Picorna': '("Caliciviridae",("Kelp fly-like group",("Aquatic Picorna-like group",("unnamed group 1",("Picornaviridae",("Fisa-like",("Nora-like",(("Iflaviridae"[&subtree="Ifla"],"Secoviridae"),("Dicistroviridae"[&subtree="Dicistro"],"unnamed group 2")))))))));'}


for taxo_group in taxonomy_trees:

    ll=bt.make_tree(taxonomy_trees[taxo_group])
    lvls=max(map(dist2root,ll.Objects))-1.9

    for k in ll.getExternal():
        k.height=lvls

    heightless=ll.getInternal(lambda k: k.height==None)
    while len(heightless)>0:
        for k in heightless:
            if len(k.children)==len([ch for ch in k.children if ch.height!=None]):
                k.height=min([ch.height-1 for ch in k.children])
        heightless=ll.getInternal(lambda k: k.height==None)

    for k in ll.Objects:
        k.length=k.height-k.parent.height

    ll.treeStats()

    background=lambda k: k.traits['bg']==True
    study=lambda k: k.traits['bg']==False


    trees={}

    labels={'Chu': 'A', 'Xinmo': 'B', 'Rhabdo-like': 'C', 'Dimarhabdo': 'D', 'CytoNucleorhabdo-like': 'E', 
            'Phlebo-like': 'A','PhasmaFeraJon': 'B', 'Orthobunya': 'C', 
            'Levi': 'A', 'Botourmia': 'B','Narna': 'C','Narna-like': 'D', 
            'Partiti-like': 'A', 
            'Flavi-like': 'A','Pesti-like': 'B', 
            'LaTardoire-like': 'A', 'MottsMills-like': 'B', 'Polero-like': 'C', 
            'Orbi-like': 'A','Cypo': 'B', 
            'LakeSinai-like': 'A', 'JingmenTombus2-like': 'B', 'Corseley-like': 'C', 
            'Omono-river-like': 'A', 'Camponotus-nipponicus-like': 'B', 'Totivirus': 'C', 'Chryso': 'D', 
            'Ifla': 'A', 'Dicistro': 'B'}
    contribution=0

    all_xs=[]
    all_ys=[]


    for tfile in sorted(glob.glob(os.path.join(base_path,'%s/*.newick'%(taxo_group))),key=lambda w: labels[os.path.basename(w).replace('.newick','')]):
        tname=os.path.basename(tfile).replace('.newick','')
        print(tname)
        subtree=bt.loadNewick(tfile)
    #     subtree.treeStats()
        subtree.traverse_tree()
        subtree=subtree.collapseBranches(collapseIf=lambda k: k.length==0.0)

        for k in subtree.getExternal(lambda w: 'CMS00' in w.numName and 'NODE' in w.numName):
            cur_node=k
            while cur_node:
                cur_node.traits['bg']=False
                cur_node=cur_node.parent

        for k in subtree.getExternal(lambda w: 'CMS00' not in w.numName and 'NODE' not in w.numName):
            cur_node=k
            while cur_node:
                cur_node.traits['bg']=True
                cur_node=cur_node.parent

        for pol in pol2contigs:
            contigs=pol2contigs[pol]
            clade=set()
            for k in subtree.getExternal(lambda w: '|'.join(w.numName.split('|')[1:]) in contigs):
                cur_node=k
                while cur_node.traits['bg']==False:
                    clade.add(cur_node)
                    cur_node=cur_node.parent

            if len(clade)>0:
                Xs[pol]=len([k for k in clade if k.branchType=='leaf'])
                Ys[pol]=sum([k.length for k in clade])
                Cs[pol]=pol2colour[pol]

        blength=subtree.getParameter('length',use_trait=False,which_branches=study)
        print(blength,sum(blength))
        contribution+=sum(blength)

        subtree.sortBranches()

        trees[tname]=subtree

    tree_order=sorted(trees,key=lambda k: labels[k])

    fig = plt.subplots(figsize=(25,25),facecolor='w')

    gs = GridSpec(2, len(trees),height_ratios=[1,3],width_ratios=[trees[tname].treeHeight+10 for tname in tree_order],hspace=0.0,wspace=0.0)
    ax = plt.subplot(gs[0,0:])

    for t,tname in enumerate(trees):
        ax2 = plt.subplot(gs[1,t])

        [ax2.spines[loc].set_visible(False) for loc in ax2.spines if loc not in []]
        ax2.tick_params(size=0,labelsize=0)

        subtree=trees[tname]

        x_attr=lambda k: k.x
        y_attr=lambda k: k.y

        label=labels[tname]

        xs=list(map(x_attr,subtree.Objects))
        ys=list(map(y_attr,subtree.Objects))

        all_xs+=xs
        all_ys+=ys

        ax2.text(0.01,0.99,label,size=20,ha='left',va='top',transform=ax2.transAxes)

        subtree.plotTree(ax2,x_attr=x_attr,y_attr=y_attr,target=background,colour_function=lambda k: 'k')
        subtree.plotTree(ax2,x_attr=x_attr,y_attr=y_attr,target=study,colour_function=lambda k: 'indianred')
        subtree.plotPoints(ax2,x_attr=x_attr,y_attr=y_attr,target=lambda k: (study(k) and k.branchType=='leaf'),colour_function=lambda k: 'indianred',marker='o')

        subtree.addText(ax2,position=lambda w: (w.x+0.1,w.y),text=lambda k: k.numName.split('|')[1],va='center',size=8+80/len(subtree.getExternal()))

        ax2.plot([0,1],[0,0],color='k',lw=2)
        if t==0:
            ax2.text(0.5,-0.5,'1 aa\nsubs/site',size=12,ha='center',va='top')

        ticksize=subtree.treeHeight/5

        ax2.set_xlim(-0.1,subtree.treeHeight+10)

    print(trees)

    x=-3.5
    y=-3
    L=1

    ax.text(0.01,0.05,'Contribution: %.4f (aa subs/site)'%(contribution),size=20,color='indianred',ha='left',va='bottom',transform=ax.transAxes)

    ll.sortBranches()
    ll.plotTree(ax)

    for k in ll.getExternal():
        name=k.numName
        s=12
        if 'subtree' in k.traits:
            print(k.traits['subtree'],k.traits['subtree'] in labels)
            name='%s (%s)'%(name,labels[k.traits['subtree']])
            s=16
            ax.scatter(k.x,k.y,s=30,facecolor='indianred',zorder=99)
            ax.scatter(k.x,k.y,s=60,facecolor='k',zorder=98)

        ax.text(k.x+0.1,k.y,name,size=s,zorder=100,va='center',ha='left')
        ax.scatter(k.x,k.y,s=20,facecolor='k',zorder=97)

    [ax.spines[loc].set_visible(False) for loc in ax.spines]


    ax.tick_params(size=0,labelsize=0)

    ax.set_xlim(0,ll.treeHeight+1)
    ax.set_ylim(0,ll.ySpan)


    plt.savefig('/Users/evogytis/Documents/manuscripts/skeeters/figures/fig7_%s.pdf'%(taxo_group),
                dpi=300,bbox_inches='tight')
    plt.savefig('/Users/evogytis/Documents/manuscripts/skeeters/figures/fig7_%s.png'%(taxo_group),
                dpi=300,bbox_inches='tight')

    plt.show()
```
