```python
import subprocess
```

```python
list_of_files_to_transfer = subprocess.Popen("aws s3 ls s3://amyk-bucket/ --recursive | grep 'Mos/' | grep -v 'bowtie_' | awk 'NF>1{print $NF}'", shell=True, stdout=subprocess.PIPE).stdout.read().decode("utf-8").split("\n")


```

```python
list_of_files_to_transfer
```

```python
files_from = ["s3://amyk-bucket/"+x for x in list_of_files_to_transfer]
```

```python
files_to = ["s3://czbiohub-mosquito/contig_quality/" + x.split("Mos")[0] + "/Mos/" + x.split("Mos")[len(x.split("Mos"))-1][1:] for x in list_of_files_to_transfer]



```

```python
files_to
```

```python
cmd = ["aws s3 cp " + files_from[i] + " " + files_to[i] for i, x in enumerate(list_of_files_to_transfer)]
```

```python
with open('copy_files.sh', 'w') as f:
    for item in cmd:
        f.write("%s\n" % item)
```
